```python
# load the csv
df = pd.read_csv("test1.csv")
# Histogram untuk Purchase Price
plt.figure(figsize=(12, 6))
plt.hist(df['Purchase Price'], bins=10, edgecolor='black')
plt.title('Purchase Price Distribution')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.show()

# Histogram untuk Current Price
plt.figure(figsize=(13, 6))
plt.hist(df['Current Price'], bins=10, edgecolor='black')
plt.title('Current Price Distribution')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.show()

# Histogram untuk Gain/Loss
plt.figure(figsize=(15, 6))
plt.hist(df['Gain/Loss'], bins=10, edgecolor='black')
plt.title('Gain/Loss Distribution')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.show()

```

```python
# load the csv
df = pd.read_csv("test1.csv")
# b. Atribut Stock Name dan Symbol dengan Dividen/Share tertinggi
top_dividend = df.sort_values('Dividend/Share', ascending=False).head(5)
print("Top Stocks by Dividend/Share:")
print(top_dividend[['Stock Name','Symbol']])
# Visualisasi dengan bar plot
plt.figure(figsize=(5, 2))
plt.bar(top_dividend['Stock Name'], top_dividend['Dividend/Share'])
plt.xlabel('Stock Name')
plt.ylabel('Dividend/Share')
plt.title('Top 5 Stocks by Dividend/Share')
plt.xticks(rotation=45)
plt.show()
```

```python
# load the csv
df = pd.read_csv("test1.csv")
# b. Atribut Stock Name dan Symbol dengan Dividen/Share tertinggi
top_market = df.sort_values('Market Value', ascending=False).head(5)
print("Top Stocks by Market Value:")
print(top_market[['Stock Name','Symbol']])
# Membuat plot batang untuk menampilkan data
plt.figure(figsize=(10, 6))
plt.bar(top_market['Stock Name'], top_market['Market Value'])
# Memberikan label pada sumbu-sumbu
plt.xlabel('Stock Name')
plt.ylabel('Market Value')
# Memberikan judul pada grafik
plt.title('Top 5 Stocks by Market Value')
# Menampilkan grafik
plt.show()
```
